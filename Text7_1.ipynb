```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

X=np.array([[3,3],[4,3],[1,1]])
Y=np.array([1,1,-1])

def solve_function(value):
    w1,w2,A,B,C,b=value[0],value[1],value[2],value[3],value[4],value[5]
    #拉格朗日
    return [w1-3*A-4*B+C,w2-3*A-3*B+C,A*(1-3*w1-3*w2-b),B*(1-4*w1-3*w2-b),C*(1+w1+w2+b),A+B-C]

solved=fsolve(solve_function,[0.5,0.5,0.5,0.5,0.5,0.5])
print(solved)

x=np.linspace(0,10,100)
#超平面
y=-x*solved[0]/solved[1]-solved[-1]/solved[1]

plt.plot(x,y)

plt.scatter(X[:-1,0],X[:-1,1],color='red')
plt.scatter(X[-1,0],X[-1,1],color='blue')

plt.xlabel('x0')
plt.ylabel('x1')

plt.show()
```
