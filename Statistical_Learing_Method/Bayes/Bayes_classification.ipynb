```python
#用贝叶斯分类算法解决鸢尾花分类

import pandas as pd
import numpy as np
import random
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB

X,y=datasets.load_iris(return_X_y=True)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
gnb=GaussianNB()
y_pred=gnb.fit(X_train,y_train).predict(X_test)
print("测试集数目为：%d 预测出错的数目为：%d "%(X_test.shape[0],(y_test!=y_pred).sum()))
print("分类的准确率为：%s"%(1-((y_test!=y_pred).sum()/X_test.shape[0])))





```

```python
import numpy as np
import pandas as pd
import random

dataSet=pd.read_csv('iris.csv',header=None)
dataSet.head()


#切分数据集
def randSpilt(dataSet,rate):
    l=list(dataSet.index)#提取索引
    random.shuffle(l)#打乱索引
    dataSet.index=l#将打乱后的索引重新复制给原始数据
    n=dataSet.shape[0]#总数据量
    m=int(n*rate)#训练集数量
    train=dataSet.loc[range(m),:]#前m行为训练集
    test=dataSet.loc[range(m,n),:]#后n-m行为测试集
    dataSet.index=range(dataSet.shape[0])#更新原数据集的索引
    test.index=range(test.shape[0])#更新测试集的索引
    return train,test

train,test=randSpilt(dataSet,0.8)

print(train.index)
print(test.index)
print(train.loc[:5,:])
print(type(train))

print(train.iloc[:,-1].value_counts())

#构建高斯朴素贝叶斯分类器
def gnb_classify(train,test):
    labels=train.iloc[:,-1].value_counts().index#提取标签种类
    mean=[]#存放每个类的均值
    std=[]#存放每个类的方差
    result=[]#存放测试集的预测结果
    
    """
    res=[]
    for i in labels:
        item=train.loc[train.iloc[:,-1]==i,:]
        
        for k in range(2):
            tmp=[]
            score=0
            for j in range(1,len(item)):
                score+=float(item.iloc[j,k])
            tmp.append(score)
        res.append(tmp)
    
    res=pd.DataFrame(res)
    print('!!!!!!!!!!!!!',res)
    """

    for i in labels:
        item=train.loc[train.iloc[:,-1]==i,:]#分别提取每一种类别
        m=item.iloc[1:,:-1].mean()#计算当前类别的均值
        s=np.sum((item.iloc[1:,:-1]-m)**2)/(item.shape[0])#计算当前类别的方差
        mean.append(m)
        std.append(s)
        
        #--------------!!!!!!!------------
        #print(m)
    means=pd.DataFrame(mean,index=labels)#变为DF式，索引为类标签
    stds=pd.DataFrame(std,index=labels)
    
    for j in range(test.shape[0]):
        iset=test.iloc[j,:-1].tolist()
        #print()
        
        iprob=np.exp(-1*(iset-means)**2/(2*stds))/(np.sqrt(2*np.pi*stds))#正态分布
        prob=1
        for k in range(test.shape[1]-1):#遍历每个特征
            prob*=iprob[k]
            cla=prob.index[np.argmax(prob.values)]#返回最大概率的类别
        result.append(cla)
    test['predict']=result
    acc=(test.iloc[:,-1]==test.iloc[:,-2]).mean()#计算预测准确率
    print(f'模型预测准确率为{acc}')
    return test
    


train,test=randSpilt(dataSet,0.8)
gnb_classify(train,test)
    
```
